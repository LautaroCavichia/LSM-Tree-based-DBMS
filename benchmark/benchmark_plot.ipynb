```python
import pandas as pd
import matplotlib.pyplot as plt

# Load benchmark results
filename = "./build/benchmark_results.csv"
df = pd.read_csv(filename)

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(df["MemtableSizeKB"], df["InsertTimeMs"], marker='o', label="Insert Time (ms)")
plt.plot(df["MemtableSizeKB"], df["ReadTimeMs"], marker='s', label="Read Time (ms)")

# Labels and legend
plt.xlabel("Memtable Size (KB)")
plt.ylabel("Time (ms)")
plt.title("LSM-Tree DB Benchmark")
plt.legend()
plt.grid()

# Show plot
plt.show()
```
